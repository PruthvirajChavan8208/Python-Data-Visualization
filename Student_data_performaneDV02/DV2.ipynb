```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv("cleaned_student_data.csv")
```

```python
ds.columns
```

```python
sns.countplot(data=ds,x="Gender")
plt.title("Total Studnet Count Gender 
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(7,4))
sns.countplot(data=ds,x="Parental_Support",palette="Set2")
plt.title("Parental_Support for students (Distribution)")
plt.show()
```

```python
sns.countplot(data=ds,x="Parental_Support",hue='Gender',palette='bright')
plt.xlabel("Parental_Support",fontsize=12)
plt.ylabel("Number Of Students",fontsize=11)
plt.title("Parental_Support split by Gender")
#plt.legend(title="Gender")
plt.show()
```

```python
sns.countplot(data=ds,x="Final_Grade",hue="Parental_Support",palette="deep")
plt.title("Parental_Support wise Performance",fontsize=16)
plt.xlabel("Final Grade")
plt.ylabel("Count of students")
plt.show()
```

```python
sns.countplot(data=ds,x="Final_Grade",hue="Gender",palette="deep")
plt.title("Gender wise Performance",fontsize=16)
plt.xlabel("Final Grade")
plt.ylabel("Count of students")
plt.show()
```

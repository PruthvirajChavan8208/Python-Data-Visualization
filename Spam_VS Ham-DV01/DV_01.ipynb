```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
ds=pd.read_csv("clean_spam_data.csv")
```

```python
ds.info()
```

```python
bar_width=0.2
plt.figure(figsize=(3,3))
spam_ham_count=ds["Category"].value_counts()
spam_ham_count.plot(kind="bar")
plt.title("Spam_VS_Ham")
plt.xlabel("spam(1)--ham(0)")
plt.ylabel("Total Count")           
```
